# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Select columns we are interested in and drop null values
df = clean_hospital[['Number of Discharges', 'Excess Readmission Ratio']].dropna()
```

## Analysing Conclusions From Above

Visually, the initial observations seem to make sense. However, we shouldn't draw conclusions from visual analysis alone. We need to verify:

1. Is there is a significant correlation between hospital capacity (number of discharges) and readmission rates?

```python
# Calculate corr coef between num discharges and readmission ratio
print('Corr coef:', np.corrcoef(df['Number of Discharges'], df['Excess Readmission Ratio'])[0, 1])
```

`It turns out the first claim is not that strong.`

2. Is there a significant difference between the `mean` `Excess Readmission Ratio` in small and large hospitals?

We will answer this below:

`Small Hospital: 0 <= num discharges < 300`

`Small Hospital: 300 <= num discharges`
## Hypothesis Test

`H0: Mu1 - Mu2 = 0`

`Ha: Mu1 - Mu2 != 0`

`α  = 0.01`

```python
# Assign small hospitals: <= 100 discharges
# Assign large hospitals: > 100 discharges
small = df[df['Number of Discharges'] < 300]
large = df[df['Number of Discharges'] >= 300]
```

```python
# Compute mean of small and large hospitals
x_bar_small = np.mean(small['Excess Readmission Ratio'])
x_bar_large = np.mean(large['Excess Readmission Ratio'])

# Compute num of small and large hospitals
num_small = len(small)
num_large = len(large)

# Compute standard error of small and large hospitals
SE_small = np.std(small['Excess Readmission Ratio'])
SE_large = np.std(large['Excess Readmission Ratio'])

# Compute z-score and p-value
z_stat = (x_bar_small - x_bar_large) / np.sqrt( ((SE_small**2)/num_small) + ((SE_large**2)/num_large) )
p_value = stats.norm.cdf(-z_stat) * 2 # Two tailed test

# Display results
print('z-stat:', z_stat)
print('p-value:', stats.norm.cdf(-z_stat) * 2)
```

With a `z-stat of ~8.2`, our `p-value is much smaller than alpha`, so `we reject the null hypothesis` and conclude that there is a significant statistical difference in the mean readmission ratio within small and large hospitals.

## Statistical vs Practical Significance

Statistically, we have found a significant difference in the mean readmission ratios within small and large hospitals. Nevertheless, both types of hospitals had a mean readdmission rate that was very close to `1`. The main concern however is to look at hospitals with `readmission ratios > ~1.2` as they pose the highest risk.
