## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Read in dataset as df
world_bank_df = pd.read_json('data/world_bank_projects.json');
```

```python
# Question 1
# Group by country, count values, sort by largest first, select first 10 values
world_bank_df.groupby('countryshortname') \
             .size() \
             .sort_values(ascending=False)[:10]
```

```python
# Load in JSON data as a string
world_bank_data = json.load((open('data/world_bank_projects.json')));
```

```python
# Normalize JSON data to extract 'mjtheme_namecode'
proj_themes_df = json_normalize(world_bank_data, 'mjtheme_namecode');
proj_themes_df.head()
```

```python
# Question 2
# Group by code, get size, sort by largest first, select first 10 elements 
proj_themes_df.groupby('code') \
              .size() \
              .sort_values(ascending=False)[:10]
```

```python
# Question 3
# Create dictionary to map code numbers to missing names in <proj_themes_df>
code_name_dict = {'1':'Economic management', '2':'Public sector governance',
                  '3':'Rule of law', '4':'Financial and private sector development',
                  '5':'Trade and integration', '6':'Social protection and risk management',
                  '7':'Social dev/gender/inclusion', '8':'Human development',
                  '9':'Urban development', '10':'Rural development',
                  '11':'Financial and private sector development'};

# Fill in missing names in <proj_themes_df> using <code_name_dict>
proj_themes_df.name = proj_themes_df.name.replace('', proj_themes_df.code.map(code_name_dict));

proj_themes_df
```
